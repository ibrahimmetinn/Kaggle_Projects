# Functions

### The function created to evaluate the other criteria according to the monthly income scale

```python
# X= any column name in the dataframe 
def mountly_income(X):
    plt.figure(figsize=(10,6))
    plt.title(str(X)+" in Mounthly Ä°ncome Statistic")
    grouped_data = df1.groupby(X)["Monthly Income"].value_counts().reset_index(name="Count")
    sns.barplot(data=grouped_data, x="Monthly Income", y="Count", hue=X)
```

### Pie chart function

```python
def pie_chart(P):
    plt.figure(figsize=(10,6))
    explode = [0.1 for _ in range(len(df1[P].unique()))]  
    palette_color = sns.color_palette('deep') 
    plt.pie(df1[P].value_counts(), labels=df1[P].unique(), colors=palette_color, 
            explode=explode, autopct='%.0f%%') 
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title("Mountly Income Statistic")
sns.barplot(x=df["Monthly Income"].value_counts().index,y=df["Monthly Income"].value_counts());
```

# Machine Learning 

```python
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
df1.columns
```

```python
df3=df1.copy()
```

```python
df3["Monthly Income"].value_counts()
```

```python
#Encoding
df3['Gender_Encoded'] = df3['Gender'].map({'Male': 1, 'Female': 0})
df3["Marital_Status_Encoded"]=df3['Marital Status'].map({'Single':0,'Married':1,"Prefer not to say":3})
df3["Occupation_Encoded"]=df3['Occupation'].map({'Student':0,'Employee':1,"Self Employeed":3,"House wife":4})
df3["EQ_Encoded"]=df3['Educational Qualifications'].map({'Uneducated':0,'School':1,"Graduate":2,"Post Graduate":3,"Ph.D":4})
df3["MI_Encoded"]=df3['Monthly Income'].map({'No Income':0,'Below Rs.10000':5000,"10001 to 25000":17500,"25001 to 50000":37500,"More than 50000":60000})
```

```python
df3["MI_Encoded"].value_counts()
```

```python
df3.describe().T
```

## DecisionTreeRegressor

```python
y=df3["MI_Encoded"]
features = ['Age','Gender_Encoded','Marital_Status_Encoded', 'Occupation_Encoded','EQ_Encoded','Family size']
X=df3[features]
## Split into validation and training data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
# Specify Model
iowa_model = DecisionTreeRegressor(random_state=1)
# Fit Model
iowa_model.fit(train_X, train_y)
```

```python
val_predictions = iowa_model.predict(val_X)
val_mae = mean_absolute_error(val_predictions, val_y)
print("Validation MAE when not specifying max_leaf_nodes: {:,.0f}".format(val_mae))
```

```python
iowa_model = DecisionTreeRegressor(max_leaf_nodes=100,random_state=1)
iowa_model.fit(train_X, train_y)
val_predictions = iowa_model.predict(val_X)
val_mae = mean_absolute_error(val_predictions, val_y)
print("Validation MAE for best value of max_leaf_nodes: {:,.0f}".format(val_mae))
```

```python
rmse = np.sqrt(val_mae)
print("Validation RMSE: {:.2f}".format(rmse))
```

## Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
# Define the model. Set random_state to 1
rf_model = RandomForestRegressor(random_state=1)

# fit your model
rf_model.fit(train_X,train_y)

# Calculate the mean absolute error of your Random Forest model on the validation data
rf_val_mae = rf_model.predict(val_X)

print("Validation MAE for Random Forest Model: {}".format(mean_absolute_error(val_y,rf_val_mae)))
```

```python
rmse1 = np.sqrt(mean_absolute_error(val_y,rf_val_mae))
print("Validation RMSE: {:.2f}".format(rmse1))
```
